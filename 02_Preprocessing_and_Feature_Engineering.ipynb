```python
# Same process for testing data

test['BsmtFin Type 1'] = test['BsmtFin Type 1'].map({np.nan: 0, 'GLQ':6, 'ALQ':5, 'BLQ':4, 'Rec':3, 'LwQ':2, 'Unf':1})
test['BsmtFin Type 2'] = test['BsmtFin Type 2'].map({np.nan: 0, 'GLQ':6, 'ALQ':5, 'BLQ':4, 'Rec':3, 'LwQ':2, 'Unf':1})
test['Bsmt Qual'] = test['Bsmt Qual'].map({np.nan: 0, 'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1})
test['Bsmt Cond'] = test['Bsmt Cond'].map({np.nan: 0, 'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1})
test['Bsmt Exposure'] = test['Bsmt Exposure'].map({np.nan: 0, 'Gd':4, 'Av':3, 'Mn':2, 'No':1})
test['Fireplace Qu'] = test['Fireplace Qu'].map({np.nan: 0, 'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1})
test['Garage Qual'] = test['Garage Qual'].map({np.nan: 0, 'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1})
test['Garage Cond'] = test['Garage Cond'].map({np.nan: 0, 'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1})
test['Pool QC'] = test['Pool QC'].map({np.nan: 0, 'Ex':4, 'Gd':3, 'TA':2, 'Fa':1})
test['Central Air'] = test['Central Air'].map({'Yes':1, 'No':0})

test['Alley'] = test['Alley'].map({np.nan:0, 'Grvl':1, 'Pave':2})
test['Garage Type'] = test['Garage Type'].map({np.nan: 0, '2Types':1, 'Attchd':2, 'Basment':3, 'BuiltIn':4, 'CarPort':5, 'Detchd':6})
test['Garage Finish'] = test['Garage Finish'].map({np.nan: 0, 'Fin':1, 'RFn':2, 'Unf':3})
test['Fence'] = test['Fence'].map({np.nan:0, 'GdPrv':1, 'MnPrv':2, 'GdWo':3, 'MnWw':4})
test['Misc Feature'] = test['Misc Feature'].map({np.nan: 0, 'Elev': 1, 'Gar2':2, 'Othr':3, 'Shed':4, 'TenC':5})

test['Exter Qual'] = test['Exter Qual'].map({'Ex': 5, 'Gd' : 4, 'TA': 3, 'Fa': 2, 'Po': 1})
test['Exter Cond'] = test['Exter Cond'].map({'Ex': 5, 'Gd' : 4, 'TA': 3, 'Fa': 2, 'Po': 1})
test['Functional'] = test['Functional'].map({'Typ':8, 'Min1':7, 'Min2':6, 'Mod':5, 'Maj1':4, 'Maj2':3, 'Sev':2, 'Sal':1})
test['Kitchen Qual'] = test['Kitchen Qual'].map({'Ex': 5, 'Gd' : 4, 'TA': 3, 'Fa': 2, 'Po': 1})
test['Heating QC'] = test['Heating QC'].map({'Ex': 5, 'Gd' : 4, 'TA': 3, 'Fa': 2, 'Po': 1})
```

## Feature Engineering

 After cleaning all of the columns I planned on working with initially, I started to engineer features from them:
 - 'Age' is a numerical value representing the time the house was last worked on and current year (2021). This 
 - 'Location' was an attempt to try and quantify the neighbourhoods the entries were listed in. The assumption was that houses with better overall quality, condition and general functionality would make neighbourhoods more sought-after. These were split into two groups, low desirability (1) and high desirability (2), before being split again into low(1), low-mid(2), high-mid(3) and high(4).
 - 'RoadRail' is a dummy column indicating whether or not a house was near or adjacent to a road or railway.
 - 'Amenities' is a dummy column indicating whether or not a house was near or adjacent to a 'possitive' amenity (Park)
 - 'Garage is a dummy column indicating whether or not a house posessed a garage.
 
 #### Interacting Features
 After intial tests of my model, I came back to add interacting features that I assumed would boost my models performance. Again, I would focus on Quality and Condition features, presence of features and their quality, the presense of multiple features, and the presense of multiple types of features.
 

```python
# property size, location desirability, nearby amenities, number of rooms, construction materials
# age and condition

# make Age variable from Year Remod
train['Age'] = 2021 - train['Year Remod/Add']

# add OverallQual, OveralCond, ExterQual_enum, 
# ExterCond_enum, Functional_enum to create Location variable
# Location variable 4 ordinal values 1-4
train['Location'] = ((train['Overall Qual']/train['Overall Qual'].mean())
                    + (train['Overall Cond']/train['Overall Cond'].mean())
                    + (train['Exter Qual']/train['Exter Qual'].mean())
                    + (train['Exter Cond']/train['Exter Cond'].mean())
                    + (train['Functional']/train['Functional'].mean()))
# SEE IF WE CAN SPLIT THIS 4 WAYS WITH A FUNCTION
def location(row):
    if row['Location'] < train['Location'].mean():
        return 1
    else:
        return 2

train['Location_temp'] = train.apply(location, axis = 1)

#Got the two means of each group from Location_temp
Loc1 = 4.630792
Loc2 = 5.354031
def location_2(row):
    if row['Location_temp'] == 1:
        if row['Location'] < Loc1:
            return 1
        else:
            return 2
    else:
        if row['Location'] > Loc2:
            return 4
        else: 
            return 3
# Make it four groups
train['Location'] = train.apply(location_2, axis = 1)
train.drop('Location_temp', axis =1, inplace = True)

# One dummy for if near a Road or Rail
train['RoadRail'] = train.apply(lambda x : 1 if x['Condition 1'] in 'Artery, Feedr, RRNn, RRAn, RRNe, RRAe' else 0,  axis= 1)

# One dummy for if near possitive Ammenities
def amenities(row):
    if row['Condition 1'] in 'PosN, PosA' or row['Condition 2'] in 'PosN, PosA':
        return 1
    else:
        return 0

train['Amenities'] = train.apply(amenities, axis = 1)

# One dummy for if Garage
train['Garage'] = train.apply(lambda x:1 if x['Garage Type'] != 0 else 0, axis =1)

# Combine Qual+Condition of variables into interaction variables
train['Exter Overall'] = train['Exter Qual']*train['Exter Cond']
train['Bsmt Overall'] = train['Bsmt Qual']*train['Bsmt Cond']
train['Garage Overall'] = train['Garage Qual']*train['Garage Cond']
#Combine house feature with quality into interaction variables

# 'Kitchen Overall', 'Fireplace Overall', 'Pool Overall', 'Bath AbvGrd', 'Multi Bsmt Type', 'Total FinBsmt SF'
train['Kitchen Overall'] = train['Kitchen AbvGr']*train['Kitchen Qual']
train['Fireplace Overall'] = train['Fireplaces']*train['Fireplace Qu']
train['Pool Overall'] = train['Pool Area']*train['Pool QC']
train['Bath AbvGrd'] = train['Full Bath']*train['Half Bath']
train['Multi Bsmt Type'] = train['BsmtFin Type 1']*train['BsmtFin Type 2']
train['Total FinBsmt SF'] = train['BsmtFin SF 1']+train['BsmtFin SF 2']
```

```python
#SAME FOR TEST SET

# property size, location desirability, nearby amenities, number of rooms, construction materials
# age and condition

# make Age variable from Year Remod
test['Age'] = 2021 - test['Year Remod/Add']

# add OverallQual, OveralCond, ExterQual_enum, 
# ExterCond_enum, Functional_enum to create Location variable
# Location variable 4 ordinal values 1-4
test['Location'] = ((test['Overall Qual']/test['Overall Qual'].mean())
                    + (test['Overall Cond']/test['Overall Cond'].mean())
                    + (test['Exter Qual']/test['Exter Qual'].mean())
                    + (test['Exter Cond']/test['Exter Cond'].mean())
                    + (test['Functional']/test['Functional'].mean()))
# SEE IF WE CAN SPLIT THIS 4 WAYS WITH A FUNCTION

test['Location_temp'] = test.apply(location, axis = 1)

#Got the two means of each group from Location_temp
Loc1 = 2.260323
Loc2 = 5.003124
def location_2(row):
    if row['Location_temp'] == 1:
        if row['Location'] < Loc1:
            return 1
        else:
            return 2
    else:
        if row['Location'] > Loc2:
            return 4
        else: 
            return 3

test['Location'] = test.apply(location_2, axis = 1)
test.drop('Location_temp', axis =1, inplace = True)

# One dummy for if near a Road or Rail
test['RoadRail'] = test.apply(lambda x : 1 if x['Condition 1'] in 'Artery, Feedr, RRNn, RRAn, RRNe, RRAe' else 0,  axis= 1)

# One dummy for if near possitive Ammenities
def amenities(row):
    if row['Condition 1'] in 'PosN, PosA' or row['Condition 2'] in 'PosN, PosA':
        return 1
    else:
        return 0

test['Amenities'] = test.apply(amenities, axis = 1)

# One dummy for if Garage
test['Garage'] = test.apply(lambda x:1 if x['Garage Type'] != 0 else 0, axis =1)

# Combine Qual+Condition of variables into interaction variables
test['Exter Overall'] = test['Exter Qual']*test['Exter Cond']
test['Bsmt Overall'] = test['Bsmt Qual']*test['Bsmt Cond']
test['Garage Overall'] = test['Garage Qual']*test['Garage Cond']
#Combine house feature with quality into interaction variables
test['Kitchen Overall'] = test['Kitchen AbvGr']*test['Kitchen Qual']
test['Fireplace Overall'] = test['Fireplaces']*test['Fireplace Qu']
test['Pool Overall'] = test['Pool Area']*test['Pool QC']
test['Bath AbvGrd'] = test['Full Bath']*test['Half Bath']
test['Multi Bsmt Type'] = test['BsmtFin Type 1']*test['BsmtFin Type 2']
test['Total FinBsmt SF'] = test['BsmtFin SF 1']+test['BsmtFin SF 2']
```
